```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

row = 256
col = 256
img = np.zeros((row,col))
img[100:105, :] = 0.5
img[:, 100:105] = 1
plt.figure(figsize=(10,4))
plt.imshow(img)
plt.show
```

```python
import numpy as np

a = np.array([[1, 4, 5],
             [-5, 8, 9],
             [6, 8, 10],
             [0, 2, 38],])

print("type:", type(a))
print("a =", "\n",a, "\n")
print("a[1] =", a[1])
print("a[1][2] =", a[1][2])
print("a[0][-1] =", a[0][-1])
```
